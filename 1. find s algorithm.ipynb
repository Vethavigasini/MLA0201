```python
def find_s(training_data):
    """
    training_data: list of tuples (attributes_list, label)
    label should be 'Yes' or 'No'
    """
    n_attrs = len(training_data[0][0])
    hypothesis = ['∅'] * n_attrs  # most specific

    for attrs, label in training_data:
        if label.lower() == 'yes':
            for i in range(n_attrs):
                if hypothesis[i] == '∅':
                    hypothesis[i] = attrs[i]
                elif hypothesis[i] != attrs[i]:
                    hypothesis[i] = '?'
        # Ignore negative examples in FIND-S

    return hypothesis


# ---------------- DEMO DATA ----------------
data = [
    (["Sunny", "Warm", "Normal", "Strong", "Warm", "Same"], "Yes"),
    (["Sunny", "Warm", "High",   "Strong", "Warm", "Same"], "Yes"),
    (["Rainy", "Cold", "High",   "Strong", "Warm", "Change"], "No"),
    (["Sunny", "Warm", "High",   "Strong", "Cool", "Change"], "Yes"),
]

final_h = find_s(data)
print("Final hypothesis:", final_h)
```
